```python
import cv2
import pytesseract

def table_detection(image):
    # Convert the image to grayscale.
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to binarize the image.
    thresholded_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY_INV)[1]

    # Find the contours of the table.
    contours = cv2.findContours(thresholded_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contours = contours[0]

    # Find the bounding box of the table.
    table_bounding_box = cv2.boundingRect(max(contours, key=cv2.contourArea))

    return table_bounding_box

def extract_table(image, table_bounding_box):
    # Crop the image to the bounding box of the table.
    cropped_image = image[table_bounding_box[1]:table_bounding_box[1] + table_bounding_box[3],
                          table_bounding_box[0]:table_bounding_box[0] + table_bounding_box[2]]

    return cropped_image

def perform_ocr(image):
    # Recognize the text in the image using OCR.
    text = pytesseract.image_to_string(image, lang='eng')

    return text

def combine_ocr_results(text, ocr_results):
    # Create a dictionary to store the OCR results.
    ocr_results = {}

    # Combine the OCR results with the normal text.
    output_text = ""

    for line in text.splitlines():
        if line.strip():
            value = ocr_results.get(line, "")
            output_text += line + " " + value + "\n"

    return output_text

if __name__ == "__main__":
    # Load the image.
    image = cv2.imread("/content/drive/MyDrive/easy_ocr/prime bank ltdn_4.jpg")

    # Detect the table in the image.
    table_bounding_box = table_detection(image)

    # Extract the table from the image.
    cropped_image = extract_table(image, table_bounding_box)

    # Perform OCR on the table.
    text = perform_ocr(cropped_image)

    # Combine the OCR results with the normal text.
    ocr_results = {}
    output_text = combine_ocr_results(text, ocr_results)

    # Print the output text.
    print(output_text)
```
